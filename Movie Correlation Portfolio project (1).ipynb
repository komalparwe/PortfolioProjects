```python
# First let's import the packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv( r'C:\Users\komal\Downloads\movies.csv')  
```

```python
#Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
```

```python
#dropping rows that contain NULL values and creating new df
newdf = df.dropna(axis=0)

```

```python
#using new df to find if now we have missing date
for col in newdf.columns:
    pct_missing = np.mean(newdf[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
```

```python
#data types for our columns
newdf.dtypes
```

```python
#change data type for columns
newdf['budget'] = newdf['budget'].astype('int64')
newdf['gross'] = newdf['gross'].astype('int64')
```

```python
newdf.head()
```

```python
#create correct year column 
newdf['yearcorrect'] = newdf['released'].str.extract(pat = '([0-9]{4})').astype(int)

```

```python
newdf.head()
```

```python
newdf = newdf.sort_values (by=['gross'], inplace=False, ascending=False)
```

```python
newdf.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Drop any duplicates
newdf['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#scatter plot with Budget Vs Gross

plt.scatter(x=newdf['budget'], y=newdf['gross'])
plt.title('Budget Vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
newdf.head()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=newdf, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#Let's start looking at correlation
newdf.corr(method='pearson') #pearson , kendall, spearman
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = newdf.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
